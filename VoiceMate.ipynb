```python
import speech_recognition as sr
import pyaudio
import pywhatkit
from gtts import gTTS
from playsound import playsound
import pandas as pd
import random
import pyttsx3
from pyttsx3.drivers import dummy
import datetime
import os

def speech(text):
    print(text)
    language = "en-US"
    output = gTTS(text=text, lang=language, slow=False, tld="com.au")
    
    output.save("./audio/ML_Project.mp3")
    playsound("./audio/ML_Project.mp3")

def get_audio():
    recorder = sr.Recognizer()
    with sr.Microphone() as source:
        speech("Say something")
        print("Listening.....")
        audio = recorder.listen(source)
    
    text = recorder.recognize_google(audio)
    print(f"You said: {text}")
    return text

#Jokes
def main():
    file_path = 'dad-a-base.csv'  # Update file path
    df = pd.read_csv(file_path, header=None)  # Read without column headers
    return df    
text = get_audio()

to_do_list = []
df = main()

#To-do list
def make_to_do_list():
    global to_do_list
    speech("Please dictate your to-do list. Say 'done' when you've listed all tasks.")
    while True:
        text = listen()
        if 'done' in text.lower():
            break
        elif text:
            to_do_list.append(text)
            speech(f"Added: {text}")

def show_to_do_list():
    global to_do_list
    if not to_do_list:
        speech("Your to-do list is empty.")
        print("To-Do List is Empty.")
    else:
        print("Your To-Do List:")
        for i, text in enumerate(to_do_list, start=1):
            print(f"{i}. {text}")
        speech("Here is your to-do list. I'll read it out for you.")
        for i, text in enumerate(to_do_list, start=1):
            speech(f"Number {i}: {text}")

def read_to_do_list():
    global to_do_list
    if not to_do_list:
        speech("Your to-do list is empty.")
    else:
        speech("Your to-do list includes the following tasks:")
        for i, text in enumerate(to_do_list, start=1):
            speech(f"Number {i}: {text}")

def add_task():
    global to_do_list
    speech("Please dictate the task you'd like to add.")
    text = listen()
    if text:
        to_do_list.append(text)
        speech(f"Task added: {text}")
        
#Countdown    
# Function to handle countdown
def countdown(t):
    while t:
        mins, secs = divmod(t, 60)
        timer = '{:02d}:{:02d}'.format(mins, secs)
        print(timer, end="\r")
        time.sleep(1)
        t -= 1
    speech("Countdown over!")
    
#Voice to text note creation
NOTE_DIRECTORY = 'voice_notes'

def create_note():
    try:
        print("Initializing speech recognizer...")
        r = sr.Recognizer()
        print("Recognizer initialized. Listening for audio...")
        with sr.Microphone() as source:
            audio = r.listen(source)
            print("Audio captured. Recognizing speech...")
            note_text = r.recognize_google(audio)
            print(f"You said: {note_text}")
            return note_text
    except Exception as e:
        print(f"Error creating note: {e}")

def save_note(note_text):
    try:
        print("Saving note to file...")
        timestamp = datetime.datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
        filename = f"{timestamp}.txt"
        if not os.path.exists(NOTE_DIRECTORY):
            print("Creating notes directory...")
            os.makedirs(NOTE_DIRECTORY)
        with open(os.path.join(NOTE_DIRECTORY, filename), 'w') as file:
            file.write(note_text)
            print(f"Note saved as {filename} in the {NOTE_DIRECTORY} directory.")
        return filename
    except Exception as e:
        print(f"Error saving note: {e}")

def read_saved_notes():
    try:
        print("Reading saved notes...")
        if not os.path.exists(NOTE_DIRECTORY):
            return ["No notes saved yet."]
        else:
            return os.listdir(NOTE_DIRECTORY)
    except Exception as e:
        print(f"Error reading notes: {e}")

def display_note(note_filename):
    try:
        with open(os.path.join(NOTE_DIRECTORY, note_filename), 'r') as file:
            note_text = file.read()
            return note_text
    except FileNotFoundError:
        return "Note not found."
    except Exception as e:
        print(f"Error displaying note: {e}")

def read_note_aloud(note_text):
    try:
        tts = gTTS(text=note_text, lang='en')
        tts.save("note.mp3")
        os.system("start note.mp3")  
    except Exception as e:
        print(f"Error reading note aloud: {e}")
        
        

if "youtube" in text.lower():
        speech("Finding that on youtube.")
        pywhatkit.playonyt(text)
elif "joke" in text.lower():
        random_index = random.randint(0, len(df) - 1)
        random_line = ' '.join(map(str, df.iloc[random_index]))  # Concatenate all values in the row
        engine = pyttsx3.init() # Initialize Text-to-Speech Engine
        engine.setProperty('rate', 125)
        engine.say(random_line)
        engine.runAndWait()
        
elif "notes" in text.lower():
        print("\nVoice Note Taker")
        print("1. Create a new note")
        print("2. View a note")
        print("3. Read a note aloud")
        choice = input("Choose an option: ")
        
        
        if choice == "1":
            print("Create a new note. Speak continuously until you say 'done'.")
            r = sr.Recognizer()
            note_text = ""
            while True:
                with sr.Microphone() as source:
                    print("Listening... Please speak.")
                    audio = r.listen(source, phrase_time_limit=None)
                    try:
                        spoken_text = r.recognize_google(audio)
                        print(f"Recognized Text: {spoken_text}")  # Debug print
                        spoken_text = spoken_text.strip().lower()  # Remove whitespace and convert to lowercase
                        print(f"Processed Text: {spoken_text}")  # Debug print
                        if spoken_text == "done":
                            print("Exiting loop...")
                            break
                        else:
                            note_text += spoken_text + " "
                    except sr.UnknownValueError:
                        print("Could not understand audio. Exiting loop...")
                        break
                    except sr.RequestError as e:
                        print(f"Error: {e}. Exiting loop...")
                        break
            if note_text:
                filename = save_note(note_text.strip())
                if filename:
                    print(f"Note saved as {filename} in the {NOTE_DIRECTORY} directory.")
            else:
                print("No note created.")

        elif choice == "2":
            notes = read_saved_notes()
            if notes and notes[0] != "No notes saved yet.":
                print("Select a note to read:")
                for i, note in enumerate(notes):
                    print(f"{i+1}. {note}")
                note_choice = input("Enter the number of your chosen note: ")
                try:
                    note_choice = int(note_choice) - 1
                    if note_choice < 0:
                        print("Invalid choice.")
                    else:
                        note_text = display_note(notes[note_choice])
                        print("\nNote Content:")
                        print(note_text)
                except (ValueError, IndexError):
                    print("Invalid choice.")
        elif choice == "3":
            notes = read_saved_notes()
            if notes and notes[0] != "No notes saved yet.":
                print("Select a note to read aloud:")
                for i, note in enumerate(notes):
                    print(f"{i+1}. {note}")
                note_choice = input("Enter the number of your chosen note: ")
                try:
                    note_choice = int(note_choice) - 1
                    if note_choice < 0:
                        print("Invalid choice.")
                    else:
                        note_text = display_note(notes[note_choice])
                        read_note_aloud(note_text)
                except (ValueError, IndexError):
                    print("Invalid choice.")
        else:
                print("Invalid choice. Please choose a valid option.")
elif "set timer" in text.lower():
                speech("For how many minutes and seconds? Example: 5 minutes 10 seconds")
                r = sr.Recognizer()
                with sr.Microphone() as source:
                    audio = r.listen(source)
                    try:
                        text = r.recognize_google(audio, language='en-US')
                        print("You said: ", text)
                        if "minute" in text and "second" in text:
                            text = text.replace("minutes", "minute").replace("seconds", "second")
                            parts = text.split("minute")
                            mins = int(parts[0].strip())
                            secs = int(parts[1].split("second")[0].strip())
                            total_seconds = (mins * 60) + secs
                            speech(f"Starting countdown for {mins} minutes {secs} seconds.")
                            threading.Thread(target=countdown, args=(total_seconds,)).start()
                        else:
                            speech("Sorry, didn't catch the time correctly. Please try again with 'X minutes Y seconds' format.")
                    except sr.UnknownValueError:
                        speech("Sorry, didn't catch that. Please try again.")
                    except sr.RequestError as e:
                        speech("Could not request results from Google Speech Recognition service; {0}".format(e))
elif "list" in text.lower():
    speech("What would you like to do? Make a list, show your list, read your list, add a task, or exit?")
    to_do_list = []
    while True:
        text = listen()
        if text:
                print(f"Recognized Command: {text}")
                if 'make to do list' in text.lower() or 'make a list' in text.lower():
                    to_do_list = []  # Reset the list if making a new one
                    make_to_do_list()
                elif 'show list' in text.lower():
                    show_to_do_list()
                elif 'read list' in text.lower():
                    read_to_do_list()
                elif 'add a task' in text.lower() or 'add task' in text.lower():
                    add_task()
                elif 'exit' in text.lower() or 'quit' in text.lower():
                    speech("Goodbye!")
                    break
                else:
                    speech("Sorry, I didn't understand that command. Please try again.")
elif "search" or "find" or "google" in text.lower():
        speech("Finding that for you")
        pywhatkit.search(text)
else:
        speech("Finding that for you")
        pywhatkit.search(text)
```
